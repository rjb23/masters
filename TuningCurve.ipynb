```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate a Gaussian tuning curve
def gaussian_tuning_curve(s, r_max, s_max, sigma):
    """
    Calculate the firing rate using a Gaussian tuning curve.
    
    Parameters:
    - s: The stimulus parameter (e.g., orientation angle)
    - r_max: The maximum firing rate
    - s_max: The stimulus value that elicits the maximum response
    - sigma: The width of the tuning curve
    
    Returns:
    - Firing rate as a function of the stimulus parameter
    """
    return r_max * np.exp(-0.5 * ((s - s_max) / sigma) ** 2)

# Function to simulate a cosine tuning curve
def cosine_tuning_curve(s, r_max, r_0, s_max):
    """
    Calculate the firing rate using a cosine tuning curve.
    
    Parameters:
    - s: The stimulus parameter (e.g., reaching angle)
    - r_max: The maximum firing rate
    - r_0: The baseline firing rate
    - s_max: The stimulus value that elicits the maximum response
    
    Returns:
    - Firing rate as a function of the stimulus parameter
    """
    return r_0 + (r_max - r_0) * np.cos(np.radians(s - s_max))

# Function to simulate a sigmoidal tuning curve
def sigmoidal_tuning_curve(s, r_max, s_half, delta_s):
    """
    Calculate the firing rate using a sigmoidal tuning curve.
    
    Parameters:
    - s: The stimulus parameter (e.g., retinal disparity)
    - r_max: The maximum firing rate
    - s_half: The stimulus value that elicits half of the maximum response
    - delta_s: The slope of the curve, controlling how quickly the firing rate changes
    
    Returns:
    - Firing rate as a function of the stimulus parameter
    """
    return r_max / (1 + np.exp((s_half - s) / delta_s))

# Generate synthetic data for demonstration
s_values = np.linspace(-90, 90, 300)  # Stimulus parameter range (e.g., angles from -90 to 90 degrees)

# Gaussian tuning curve parameters
r_max_gaussian = 50  # Maximum firing rate in Hz
s_max_gaussian = 0   # Preferred stimulus (angle)
sigma_gaussian = 20  # Width of the tuning curve

# Cosine tuning curve parameters
r_max_cosine = 55   # Maximum firing rate in Hz
r_0_cosine = 30     # Baseline firing rate in Hz
s_max_cosine = 45   # Preferred stimulus (angle)

# Sigmoidal tuning curve parameters
r_max_sigmoidal = 60  # Maximum firing rate in Hz
s_half_sigmoidal = 10 # Stimulus value at half-max response
delta_s_sigmoidal = 15 # Slope of the curve

# Compute firing rates for each tuning curve
firing_rate_gaussian = gaussian_tuning_curve(s_values, r_max_gaussian, s_max_gaussian, sigma_gaussian)
firing_rate_cosine = cosine_tuning_curve(s_values, r_max_cosine, r_0_cosine, s_max_cosine)
firing_rate_sigmoidal = sigmoidal_tuning_curve(s_values, r_max_sigmoidal, s_half_sigmoidal, delta_s_sigmoidal)

# Plotting the tuning curves
plt.figure(figsize=(12, 8))

# Plot Gaussian tuning curve
plt.subplot(3, 1, 1)
plt.plot(s_values, firing_rate_gaussian, color='blue')
plt.title('Gaussian Tuning Curve')
plt.xlabel('Stimulus (s)')
plt.ylabel('Firing Rate (Hz)')

# Plot Cosine tuning curve
plt.subplot(3, 1, 2)
plt.plot(s_values, firing_rate_cosine, color='green')
plt.title('Cosine Tuning Curve')
plt.xlabel('Stimulus (s)')
plt.ylabel('Firing Rate (Hz)')

# Plot Sigmoidal tuning curve
plt.subplot(3, 1, 3)
plt.plot(s_values, firing_rate_sigmoidal, color='red')
plt.title('Sigmoidal Tuning Curve')
plt.xlabel('Stimulus (s)')
plt.ylabel('Firing Rate (Hz)')

# Display the plots
plt.tight_layout()
plt.show()
```
